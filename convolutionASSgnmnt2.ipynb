```python
#Importing necessary modules..
import numpy as np
import pandas as pd
import keras
import skimage.measure 
from keras.layers import Conv2D
import matplotlib.pyplot as plt
from keras.models import Sequential

```

```python
#Importing the image
from PIL import Image
image=Image.open('/content/drive/MyDrive/istockphoto-1026205392-612x612.jpg')
plt.imshow(image)
```

```python
#resizing
rsize = image.resize((np.array((25,25))).astype(int))
image = np.asarray(rsize)
image.shape
```

```python
#Puting the image into an array
x=image.reshape(1,25,25,3)
```

**Terminologies**:

**Kernel Size:** The kernel size defines the field of view of the convolution. A common choice for 2D is 3  that is 3x3 pixels.


**Stride:** The stride defines the step size of the kernel when traversing the image. While its default is usually 1, we can use a stride of 2 for downsampling an image similar to MaxPooling.


**Padding:** The padding defines how the border of a sample is handled. A (half) padded convolution will keep the spatial output dimensions equal to the input, whereas unpadded convolutions will crop away some of the borders if the kernel is larger than 1.

***Below is an pictorial representation of 2D convolution using a kernel size of 3, stride of 1 and padding.***

1_1okwhewf5KCtIPaFib4XaA.gif

```python
def Convolution(x,kernel_size,strides=1, padding='valid'):

 model=Sequential()

 model.add(Conv2D(1, kernel_size=kernel_size,strides=strides,padding=padding,input_shape=(25,25,3)))
 yhat = model.predict(x)

 y=yhat.reshape(yhat.shape[1],yhat.shape[2])
 filtered_image = Image.fromarray(y)
 
 return filtered_image
```

```python
Filtered_image_11=Convolution(x,(3,3))
Filtered_image_12=Convolution(x,(5,5))
Filtered_image_21=Convolution(x,(3,3),2)
Filtered_image_22=Convolution(x,(5,5),2)
Filtered_image_31=Convolution(x,(3,3),1,'same')
Filtered_image_32=Convolution(x,(5,5),1,'same')
```

**Applying filter size** (3*3)

```python
#plotting image and finding entropy...
plt.imshow(Filtered_image_11)
entropy1 = skimage.measure.shannon_entropy(Filtered_image_11)
print("Entropy:",entropy1)

```

**filter size** (5*5)


```python
plt.imshow(Filtered_image_12)
entropy2 = skimage.measure.shannon_entropy(Filtered_image_12)
print("Entropy:",entropy2)
```

```python
plt.imshow(Filtered_image_21)
entropy3 = skimage.measure.shannon_entropy(Filtered_image_21)
print("Entropy:",entropy3)
```

```python
plt.imshow(Filtered_image_22)
entropy4 = skimage.measure.shannon_entropy(Filtered_image_22)
print("Entropy:",entropy4)
```
